```python
from multiprocessing import Process
```

```python
def foo(a: str="hug"):
    print(a)
```

```python
import time
t0 = time.perf_counter()

job_vars = ["quick", "brown", "fox", "jumped"]
processes = [Process(target=foo, kwargs=dict(a=var)) for var in job_vars]

for process in processes:
    process.start()
    
print(f"Execution time: {time.perf_counter() - t0}")
```

```python
def make_square(side_length):
    square = []
    int_set = [x for x in range(1, side_length**2 + 1)]
    for index in range(side_length):
        start = side_length*index
        square.append(int_set[start: start + side_length])
    return np.array(square)
```

```python
af = make_square(11)
print(af)
```
